```python
from joblib import dump, load

scaler=load(scaler_path)
```

```python
def surrogate_evaluation(x1,x2,x3,x4,x5,x6,output):
  if output == 'c':
    surrogate_model = surrogate_model_c
  else:
    surrogate_model = surrogate_model_r
  matriz = np.zeros((1,6))
  matriz[0,0] = x1
  matriz[0,1] = x2
  matriz[0,2] = x3
  matriz[0,3] = x4
  matriz[0,4] = x5
  matriz[0,5] = x6
  scaled_matriz  = scaler.transform(matriz)
  x1 = scaled_matriz[0,0]
  x2 = scaled_matriz[0,1]
  x3 = scaled_matriz[0,2]
  x4 = scaled_matriz[0,3]
  x5 = scaled_matriz[0,4]
  x6 = scaled_matriz[0,5]
  x_params = np.array([x1, x2, x3, x4, x5, x6, 1])
  x_params = np.expand_dims(x_params, axis = 0)
  hf_mean_surrogate_model, hf_var_surrogate_model = surrogate_model.predict(x_params)
  Y = hf_mean_surrogate_model
  return Y
```

```python
# we consider a rango of 30% around the mean

rango = 0.5

# Define the model inputs
problem = {
    'num_vars': 6,
    'names': ['c', 'beta', 'c1', 'c3', 'k', 'Ks'],
    'bounds': [[-rango*356.7 + 356.7 , 356.7 + rango*356.7],
               [-rango*1.075 + 1.075 , 1.075 + rango*1.075],
               [-rango*278.2 + 278.2 , 278.2 + rango*278.2],
               [-rango*5.766 + 5.766 , 5.766 + rango*5.766],
               [-rango*10000 + 10000 , 10000 + rango*10000],
               [-rango*0.08 + 0.08   , 0.08  + rango*0.08]]
}
```

```python
# Generate samples
param_values = saltelli.sample(problem, 1024)

# Run model
Y_sobol_C = np.array(list(map(lambda p: surrogate_evaluation(p[0], p[1], p[2], p[3], p[4], p[5], output='c'), param_values)))
Y_sobol_C = Y_sobol_C[:,0][:,0]
Y_sobol_R = np.array(list(map(lambda p: surrogate_evaluation(p[0], p[1], p[2], p[3], p[4], p[5], output='r'), param_values)))
Y_sobol_R = Y_sobol_R[:,0][:,0]

# Perform analysis
print("Resistance Sobol Indices")
R_Si_hf_gp = sobol.analyze(problem, Y_sobol_R, print_to_console = True) # Sobol indices for Resistance (R)
print("")
print("Compliance Sobol Indices")
C_Si_hf_gp = sobol.analyze(problem, Y_sobol_C, print_to_console = True) # Sobol indices for Compliance (C)

# Plot the Sobol indices
R_Si_hf_gp.plot()
C_Si_hf_gp.plot()
```

```python
N_samples_list = np.arange(10,10000,100)
#noise_level = 0.01
samples = 1000
Mean_list_c = []
Mean_list_beta = []
Mean_list_c1 = []
Mean_list_c3 = []
Mean_list_k = []
Mean_list_Ks = []

for N_samples in N_samples_list:

  Mean_list_int_c = []
  Mean_list_int_beta = []
  Mean_list_int_c1 = []
  Mean_list_int_c3 = []
  Mean_list_int_k = []
  Mean_list_int_Ks = []

  for i in range(samples):
    
    noise_list = [0.3]

    #datos = np.zeros((5,5))
    i = 0

    for noise_level in noise_list:
      mu_c = 356.7
      c_s = np.random.uniform(-noise_level*mu_c+mu_c,mu_c+noise_level*mu_c,N_samples)

      mu_beta = 1.075
      beta_s = np.random.uniform(-noise_level*mu_beta+mu_beta,mu_beta+noise_level*mu_beta,N_samples)

      mu_c1 = 278.2
      c1_s = np.random.uniform(-noise_level*mu_c1+mu_c1,mu_c1+noise_level*mu_c1,N_samples)

      mu_c3 = 5.766
      c3_s = np.random.uniform(-noise_level*mu_c3+mu_c3,mu_c3+noise_level*mu_c3,N_samples)

      mu_k = 10000
      k_s = np.random.uniform(-noise_level*mu_k+mu_k,mu_k+noise_level*mu_k,N_samples)

      mu_Ks = 0.08
      Ks_s = np.random.uniform(-noise_level*mu_Ks+mu_Ks,mu_Ks+noise_level*mu_Ks,N_samples)

      mean_parametros_c_s = np.mean(c_s)
      mean_parametros_beta_s = np.mean(beta_s)
      mean_parametros_c1_s = np.mean(c1_s)
      mean_parametros_c3_s = np.mean(c3_s)
      mean_parametros_k_s = np.mean(k_s)
      mean_parametros_Ks_s = np.mean(Ks_s)

      Mean_list_int_c.append(mean_parametros_c_s)
      Mean_list_int_beta.append(mean_parametros_beta_s)
      Mean_list_int_c1.append(mean_parametros_c1_s)
      Mean_list_int_c3.append(mean_parametros_c3_s)
      Mean_list_int_k.append(mean_parametros_k_s)
      Mean_list_int_Ks.append(mean_parametros_Ks_s)
  
  Mean_list_c.append(Mean_list_int_c)
  Mean_list_beta.append(Mean_list_int_beta)
  Mean_list_c1.append(Mean_list_int_c1)
  Mean_list_c3.append(Mean_list_int_c3)
  Mean_list_k.append(Mean_list_int_k)
  Mean_list_Ks.append(Mean_list_int_Ks)

arraymean_c = np.asarray(Mean_list_c)
arraymean_beta = np.asarray(Mean_list_beta)
arraymean_c1 = np.asarray(Mean_list_c1)
arraymean_c3 = np.asarray(Mean_list_c3)
arraymean_k = np.asarray(Mean_list_k)
arraymean_Ks = np.asarray(Mean_list_Ks)

sigma = 1/12

# theoretical 95% confidence interval according to central limit theorem
plt.fill_between(N_samples_list,mu_c - 1.96*sigma/np.sqrt(N_samples_list),mu_c + 1.96*sigma/np.sqrt(N_samples_list), color = '0.8')

for n, pis in zip(N_samples_list,arraymean_c):
  plt.scatter([n]*pis.shape[0], pis, color = 'k', s = 1)
plt.hlines(mu_c, 0, N_samples_list[-1], colors = 'r')
plt.show()

plt.fill_between(N_samples_list,mu_beta - 1.96*sigma/np.sqrt(N_samples_list),mu_beta + 1.96*sigma/np.sqrt(N_samples_list), color = '0.8')
for n, pis in zip(N_samples_list,arraymean_beta):
  plt.scatter([n]*pis.shape[0], pis, color = 'k', s = 1)
plt.hlines(mu_beta, 0, N_samples_list[-1], colors = 'r')
plt.show()

plt.fill_between(N_samples_list,mu_c1 - 1.96*sigma/np.sqrt(N_samples_list),mu_c1 + 1.96*sigma/np.sqrt(N_samples_list), color = '0.8')
for n, pis in zip(N_samples_list,arraymean_c1):
  plt.scatter([n]*pis.shape[0], pis, color = 'k', s = 1)
plt.hlines(mu_c1, 0, N_samples_list[-1], colors = 'r')
plt.show()

plt.fill_between(N_samples_list,mu_c3 - 1.96*sigma/np.sqrt(N_samples_list),mu_c3 + 1.96*sigma/np.sqrt(N_samples_list), color = '0.8')
for n, pis in zip(N_samples_list,arraymean_c3):
  plt.scatter([n]*pis.shape[0], pis, color = 'k', s = 1)
plt.hlines(mu_c3, 0, N_samples_list[-1], colors = 'r')
plt.show()

plt.fill_between(N_samples_list,mu_k - 1.96*sigma/np.sqrt(N_samples_list),mu_k + 1.96*sigma/np.sqrt(N_samples_list), color = '0.8')
for n, pis in zip(N_samples_list,arraymean_k):
  plt.scatter([n]*pis.shape[0], pis, color = 'k', s = 1)
plt.hlines(mu_k, 0, N_samples_list[-1], colors = 'r')
plt.show()

plt.fill_between(N_samples_list,mu_Ks - 1.96*sigma/np.sqrt(N_samples_list),mu_Ks + 1.96*sigma/np.sqrt(N_samples_list), color = '0.8')
for n, pis in zip(N_samples_list,arraymean_Ks):
  plt.scatter([n]*pis.shape[0], pis, color = 'k', s = 1)
plt.hlines(mu_Ks, 0, N_samples_list[-1], colors = 'r')
plt.show()
```
